#Mean, Median, and Mode in Machine Learning:
1. Mean: The mean, also known as average, is the sum of all values in a dataset divided by the total number of values. It represents the central tendency of the data.
2. Median: The median is the middle value of a dataset when it is ordered from least to greatest. If there's an even number of values, it's the average of the two middle values. It's a robust measure of central tendency and less affected by outliers compared to the mean.
3. Mode: The mode is the value that appears most frequently in a dataset. A dataset can have one mode, more than one mode (multimodal), or no mode if all values occur with the same frequency.

In conclusion:
* Mean - The average value
* Median - The mid point value
* Mode - The most common value

###**Outlier**-
* An outlier is a data point that is significantly different from the average value of a group of statistics.
* For example, in a graph with dots, an outlier would be a dot that is far away from the other dots.
* An outlier is someone or something that is very different from the other people or things in a group.
* He hadn't stood out as an obvious outlier within the party. The company is an extreme outlier in terms of both its pricing and behaviour.

###**Central tendency**-
It is defined as “the statistical measure that identifies a single value as representative of an entire distribution.” It aims to provide an accurate description of the entire data. It is the single value that is most typical/representative of the collected data.

```python
import numpy as np
from scipy import stats

# Generate sample data
data = [15, 18, 10, 23, 18, 16, 21, 16, 19, 18]

# Mean calculation
mean = np.mean(data)

# Median calculation
median = np.median(data)

# Mode calculation
mode = stats.mode(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

###Median- If there are two numbers in the middle, divide the sum of those numbers by two.

#4. Standard Deviation:
* Standard deviation is a number that describes how spread out the values are.

* It measures the average distance of data points from the mean.

* Standard deviation is the square root of the variance.

* A low standard deviation means that most of the numbers are close to the mean (average) value.

* A high standard deviation means that the values are spread out over a wider range. A higher standard deviation indicates more dispersion in the data.

```python
import numpy
speed = [86,87,88,86,87,85,86]
mean = np.mean(speed)
print("Mean:", mean)
```

```python
import numpy

speed = [86,87,88,86,87,85,86]

x = numpy.std(speed)

print(x)
```

Meaning that most of the values are within the range of 0.9 from the mean value, which is 86.4.

#5. Variance:

* Variance measures the average squared deviation of each data point from the mean of the dataset.
* It indicates how much the data points vary from the mean.
* In fact, if you take the square root of the variance, you get the standard deviation!
* Or the other way around, if you multiply the standard deviation by itself, you get the variance!

```python
import numpy as np

# Sample data
data = [10, 12, 14, 16, 18]

# Variance calculation
variance = np.var(data)

# Standard deviation calculation
std_deviation = np.sqrt(variance)

print("Variance:", variance)
print("Standard Deviation:", std_deviation)
```

* Standard deviation and variance are two key measures commonly used in the financial sector.
* Standard deviation is the spread of a group of numbers from the mean.
* The variance measures the average degree to which each point differs from the mean.
* While standard deviation is the square root of the variance, variance is the average of the squared difference of each data point from the mean.
* The two concepts are useful and significant for traders, who use them to measure market volatility.

```python
from itertools import permutations, combinations

# Sample data
data = [1, 2, 3]

# Permutations
print("Permutations:")
for perm in permutations(data):
    print(perm)

# Combinations
print("\nCombinations:")
for r in range(1, len(data) + 1):
    for comb in combinations(data, r):
        print(comb)
```

####Python’s Itertool is a module that provides various functions that work on iterators to produce complex iterators. This module works as a fast, memory-efficient tool that is used either by themselves or in combination to form iterator algebra.

* Permutations: We use permutations() function from the itertools library to generate all possible permutations of the sample data. Each permutation represents a unique arrangement of the elements.
* Combinations: We use combinations() function from the itertools library to generate all possible combinations of the sample data. We iterate over different values of r (the size of the combinations) to generate combinations of different lengths.

# Bayes' Theorem in Machine Learning:
* Bayes' theorem describes the probability of an event, based on prior knowledge of conditions that might be related to the event.
* It is based on Conditional probablity.
* In probablity theory, it relates the conditional probablity & marginal probablities of a random event.
* In machine learning, Bayes' theorem is often used in the context of Bayesian inference, where it enables us to update our beliefs about the probability of a hypothesis given new evidence.
* Bayes' theorem is a fundamental concept in probability theory, widely used in machine learning for classification, anomaly detection, and probabilistic inference tasks.

* P(A|B) is the probability of event A when event B happens

* Conditional probability is one of the types of probability in probability theory, where the probability of one event is dependent on the other event already happened.
* Marginal Probablity- In machine learning, marginal probability is the probability of an event for one random variable, regardless of the outcome of another random variable.

```python
# Define probabilities
p_disease = 0.01  # Prior probability of the disease
p_positive_given_disease = 0.99  # Probability of a positive test given the patient has the disease
p_positive_given_no_disease = 0.05  # Probability of a positive test given the patient doesn't have the disease

# Calculate probability of having the disease given a positive test using Bayes' theorem
p_positive = p_positive_given_disease * p_disease + p_positive_given_no_disease * (1 - p_disease)
p_disease_given_positive = (p_positive_given_disease * p_disease) / p_positive

print("Probability of having the disease given a positive test:", p_disease_given_positive)
```

###Explanation:

* **p_disease**: This represents the prior probability of a patient having the disease. It's the probability of a random patient having the disease before any diagnostic test is performed.
* **p_positive_given_disease**: This represents the probability of getting a positive test result given that the patient has the disease.
* p_positive_given_no_disease: This represents the probability of getting a positive test result given that the patient doesn't have the disease.

* **p_positive**: This calculates the total probability of getting a positive test result. It's computed by adding the probability of getting a positive result given the patient has the disease (p_positive_given_disease) and the probability of getting a positive result given the patient doesn't have the disease (p_positive_given_no_disease), weighted by the respective probabilities of having and not having the disease.
* **p_disease_given_positive**: This applies Bayes' theorem to calculate the probability of the patient having the disease given a positive test result. It's computed by multiplying the prior probability of the disease (p_disease) with the probability of getting a positive test result given the patient has the disease (p_positive_given_disease), and then dividing by the total probability of getting a positive test result (p_positive).
